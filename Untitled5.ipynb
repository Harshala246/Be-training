```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Iris cvs.csv')
df
```

```python
plt.bar(df['Species'],df['PetalLengthCm'])
plt.grid()
```

```python
plt.barh(df['Species'], df['PetalLengthCm'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Species')
plt.title('Petal Length by Species')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
labels = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

sizes = [df['SepalLengthCm'].mean(), df['SepalWidthCm'].mean(), df['PetalLengthCm'].mean(), df['PetalWidthCm'].mean()]
colors = ['silver', 'green', 'pink', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.f%%', startangle=140)

plt.title('pie chart')

plt.show()
```
